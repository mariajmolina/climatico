## Figure 12 (Journal of Climate submission; Molina et al.)

December-February (DJF) mean vertical temperature (◦C) to a depth of 285 meters during (a) El Niño(DJF ONI≥+0.5) and (g) La Niña (DJF ONI≤-0.5) shown for the CESM1 control (years 800-1599) acrossthe Niño 3.4 region. DJF mean vertical temperature anomalies during El Niño and La Niña are also shown forsensitivity experiments (b-f, h-l) as computed from the respective CESM1 control (a,g). Time periods consideredas in Fig. 11. Dashed lines represent 20◦C (black), 15◦C (dark grey), and 12◦C (grey) isotherms.

**Figure by: Maria J. Molina, NCAR**

```python
# imports 

import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
from climatico import enso
```

```python
from config import directory_figs, directory_data
```

```python
def pop_lon_indx():
    """
    Extract mask for the pacific slab region. Mask contains ones and nans.
    """
    for_lon = xr.open_dataset('/glade/scratch/molina/amoc_exp/b.e11.B1850LENS.f09_g16.FWAtSalG02Sv.pop.h.SST.000101-005012.nc')
    mask = for_lon['SST'].where((for_lon['TLAT']<5) & (for_lon['TLAT']>-5) & (for_lon['TLONG']>165) & (for_lon['TLONG']<-90+360), 
                                 drop=False).isel(z_t=0, time=0).values
    return np.where(np.isnan(mask), np.nan, 1)

def compute_iso(threedim_array, mask):
    """
    Create array of depth of isotherm using 3d iso array and 2d mask.
    Args:
        threedim_array (numpy array): Isotherm values.
        mask (numpy array): Mask from pop_lon_indx.
    Returns:
        One dimensional array across Pacific slab region.
    """
    newmask = np.nanmean(np.nanmean(threedim_array[:,:,:] * mask[np.newaxis,:,:], axis=0), axis=0) * 0.01
    return newmask[~np.isnan(newmask)]
```

```python
nino = xr.open_dataset(f'{directory_data}ensoslabs_ninodata.nc')
nina = xr.open_dataset(f'{directory_data}ensoslabs_ninadata.nc')
nino_iso20 = xr.open_dataset(f'{directory_data}ninoslabs_DEPTH_OF_20C.nc')
nina_iso20 = xr.open_dataset(f'{directory_data}ninaslabs_DEPTH_OF_20C.nc')
nino_iso15 = xr.open_dataset(f'{directory_data}ninoslabs_DEPTH_OF_15C.nc')
nina_iso15 = xr.open_dataset(f'{directory_data}ninaslabs_DEPTH_OF_15C.nc')
nino_iso12 = xr.open_dataset(f'{directory_data}ninoslabs_DEPTH_OF_12C.nc')
nina_iso12 = xr.open_dataset(f'{directory_data}ninaslabs_DEPTH_OF_12C.nc')
```

```python
# generate climatology values
TEMP_cntrl = 'b.e11.B1850C5CN.f09_g16.005.pop.h.TEMP.*.nc'
nino_TEMP = enso.DefineNino(nino='nino34', cutoff=0.5, runningmean=3)
ds_ = xr.open_mfdataset(f'{directory_data}{TEMP_cntrl}',
                        combine='by_coords',
                        preprocess=nino_TEMP.nino)
ds_ = ds_['TEMP'].mean(dim=['nlat'], skipna=True).sel(z_t=slice(500., 300. * 100.)).resample(time='QS-DEC').mean(skipna=True)
ds_ = ds_[(ds_['time'].dt.month==12)]
ds_ = ds_[(ds_['time'].dt.year>=801)&(ds_['time'].dt.year<=1599)]
ds_  = ds_.mean(dim='time', skipna=True).values
```

```python
# labels
est_lons = nino['TLONG'].mean(dim=['nlat'], skipna=True).values-360
depth    = nino.coords['z_t'].values
```
